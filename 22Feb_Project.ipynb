```python
import pandas as pd
import numpy as np

# Function to detect outliers using IQR
def detect_outliers_iqr(df, columns):
    outliers = {}
    for col in columns:
        # Ensure that the column contains numeric data
        df[col] = pd.to_numeric(df[col], errors='coerce')  # Coerce errors to NaN

        # Drop rows with NaN values in the column
        valid_data = df[col].dropna()

        # Apply IQR method if the column has enough valid data
        if len(valid_data) > 0:
            Q1 = valid_data.quantile(0.25)
            Q3 = valid_data.quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR

            # Identify outliers
            outliers[col] = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    
    return outliers

# Ensure you have the correct numerical columns
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns

# Get outliers for numerical columns
outliers = detect_outliers_iqr(df, numerical_columns)
```

```python
# Number of boxplots per row
plots_per_row = 5
plt.figure(figsize=(20, len(outliers) // plots_per_row * 4 + 4))  
for i, col in enumerate(outliers):
    plt.subplot((len(outliers) + plots_per_row - 1) // plots_per_row, plots_per_row, i + 1)
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot for {col}")
plt.tight_layout()
plt.show()
```

```python
## Distribution of numerical columns
plt.figure(figsize=(18, 12))
plt.subplot(2, 3, 1)
sns.histplot(df['Predicted_Stockout_Risk'], kde=True, color='midnightblue')
plt.title('Distribution of Predicted_Stockout_Risk')

plt.subplot(2, 3, 2)
sns.histplot(df['Daily_Consumption'], kde=True, color='midnightblue')
plt.title('Distribution of Daily Consumption')
```

```python
plots_per_row = 5
def remove_outliers(df, columns):
    df_clean = df.copy()
    for col in columns:
        Q1 = df_clean[col].quantile(0.25)
        Q3 = df_clean[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df_clean = df_clean[(df_clean[col] >= lower_bound) & (df_clean[col] <= upper_bound)]
    return df_clean

# Apply outlier removal
df_clean = remove_outliers(df, outliers)

plt.figure(figsize=(20, len(outliers) // plots_per_row * 4 + 4))  
for i, col in enumerate(outliers):
    plt.subplot((len(outliers) + plots_per_row - 1) // plots_per_row, plots_per_row, i + 1)
    sns.boxplot(x=df_clean[col])
    plt.title(f"Boxplot for {col} (Outliers Removed)")
plt.tight_layout()
plt.show()
```
