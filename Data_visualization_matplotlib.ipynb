```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,3,5,7,9]
plt.plot(x,y,linestyle='dashed',color='blue',marker=(5,1),linewidth=1.5)
plt.title("Line plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[1000,3000,2000,500,800,1500]
plt.plot(x,y,linestyle='dashed',color='green',marker='o')
plt.title("Growth of savings")
plt.xlabel("Months")
plt.ylabel("Savings amount")
plt.show()
```

```python
categories=['A','B','C','D','E']
values=[10,15,7,12,9]
plt.bar(categories,values,color='skyblue',edgecolor='black')
plt.title("Bar chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
categories=['A','B','C','D','E']
values=[10,15,7,12,9]
plt.bar(categories,values,color=['lavender','lightgreen','skyblue','lightpink','lightgrey'],edgecolor='black')
plt.title("Bar chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
ages=[30,20,18,15,19,35,38,27,45,43,50,73,82,44,12,87,28,66,68]
b=[10,15,20,25,30,35,40]

plt.hist(ages,bins=b,color='lavender',edgecolor='black')
#bins define the intervals of grouping the data
plt.title("Age Distribution of customers")
plt.xlabel("Age Ranges")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
languages=['python','java','ruby','c++']
percentages=[40,30,20,10]
plt.pie(percentages,labels=languages,colors=['lightblue','indigo','lightpink','lightgreen'],autopct='%1.1f%%',startangle=90)
plt.title('Programming languages usage')
plt.legend()
plt.show()
```

```python
x=[10,20,30,40,50]
y=[10,15,20,25,30]
#plt.plot(x,y)
plt.scatter(x,y,color='brown',s=80,edgecolor='black')
plt.title("Scatter plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```
