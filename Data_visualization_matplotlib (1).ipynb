```python
x=[1,2,3,4,5]
y=[1,3,5,7,9]
plt.plot(x,y,linestyle='dashed',color='blue',marker=(5,1),linewidth=1.5)
plt.title("Line plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
categories=['A','B','C','D','E']
values=[10,15,7,12,9]
plt.bar(categories,values,color='skyblue',edgecolor='black')
plt.title("Bar chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
categories=['A','B','C','D','E']
values=[10,15,7,12,9]
plt.bar(categories,values,color=['lavender','lightgreen','skyblue','lightpink','lightgrey'],edgecolor='black')
plt.title("Bar chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
ages=[30,20,18,15,19,35,38,27,45,43,50,73,82,44,12,87,28,66,68]
b=[10,15,20,25,30,35,40]

plt.hist(ages,bins=b,color='lavender',edgecolor='black')
#bins define the intervals of grouping the data
plt.title("Age Distribution of customers")
plt.xlabel("Age Ranges")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
languages=['python','java','ruby','c++']
percentages=[40,30,20,10]
plt.pie(percentages,labels=languages,colors=['lightblue','indigo','lightpink','lightgreen'],autopct='%1.1f%%',startangle=90)
plt.title('Programming languages usage')
plt.legend()
plt.show()
```

```python
#Sales Analysis

#Use line plots to track monthly sales and histograms for order quality distribution.
#Customer Demographics: Visualize age distribution(histogram) and spending patterns (scatter plot)
#School Performance: Compare student's marks across subjects (bar chart) and identify outliers(boxplot).
```

```python
#Sales Analysis
months=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
sales=[2800,3000,2500,4200,3800,1900,2200,3950,4350,2000,5000,5800]
plt.plot(months,sales,color='blue',marker='o')
plt.title("Monthly Sales Analysis")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.grid()
plt.show()
```

```python
#Customer Demographics:
#Age distribution
ages=[50,20,28,15,19,35,38,27,45,43,50,73,82,44,12,87,28,66,68,18,16,56,85]
b=[10,20,30,40,50,60,70,80,90]

plt.hist(ages,bins=b,color='lightgreen',edgecolor='black')
#bins define the intervals of grouping the data
plt.title("Age Distribution of customers")
plt.xlabel("Age Ranges")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
#Customer Spending Patterns by Age
ages=[30,20,18,15,19,35,38,27,45,43,50,73,82,44,66]
spendings=[1800,2200,1500,3000,4100,1200,5200,4600,2500,3100,4400,3600,4500,3300,7000]
plt.scatter(ages,spendings, color='blue', alpha=0.6)
plt.title('Customer Spending Patterns by Age')
plt.xlabel('Age')
plt.ylabel('Spendings ($)')
plt.grid()
plt.show()
```

```python
#Student's marks over all subjects
import matplotlib.pyplot as plt
subjects = ['Mathematics', 'Chemistry', 'Physics', 'History', 'Hindi', 'English', 'Biology']
marks = [79, 88, 80, 91, 95, 97, 99]
plt.bar(subjects, marks, color=['indigo', 'lavender', 'lightgreen', 'skyblue', 'lightpink', 'blue', 'lightgrey'], edgecolor='black')
plt.title("Student's Performance")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.ylim(70, 100) 
plt.show()
```

```python
#Box plot
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Student': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'],
    'Math': [85, 90, 78, 88, 92, 95, 100, 60, 70, 110], 
    'Science': [80, 85, 90, 95, 100, 105, 110, 115, 120, 125],
    'English': [70, 75, 80, 85, 90, 95, 100, 105, 110, 115],
    'History': [60, 65, 70, 75, 80, 85, 90, 95, 100, 105],
    'Art': [90, 92, 94, 96, 98, 100, 102, 104, 106, 108]
}
df = pd.DataFrame(data)
df_melted = df.melt(id_vars='Student', var_name='Subject', value_name='Marks')
plt.figure(figsize=(10, 6))
sns.boxplot(x='Subject', y='Marks', data=df_melted)
plt.title('Boxplot of Student Marks Across Subjects')
plt.ylabel('Marks')
plt.ylim(0, 130) 
plt.grid(axis='y')
plt.show()
```
